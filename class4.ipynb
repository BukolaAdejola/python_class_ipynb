```python
# Matplotlib - there is no pie chart in seaborn, that is why we have to go through creating data in 
#plt first to build a pie chart

temp_data = rent_data[['City', 'Rent']].groupby(['City']).sum()
temp_data.reset_index(inplace=True)
plt.pie(data=temp_data, x='Rent', labels='City')
plt.title('Pie Chart of rent by city')
```

## Multiple Plots

```python
# using Matplotlib to customize chart size
```

```python
size = plt.figure(figsize=(20, 10))
fig, axes = plt.subplots(2, 1, figsize=(20, 10))

sns.countplot(data=rent_data, x='City', ax=axes[0])
sns.countplot(data=rent_data, x='Point of Contact', ax=axes[1])

# set Title on each chart
axes[1].set_title('No of listing in each city')
axes[0].set_title('listing of each contact')

# set Y label
axes[1].set_ylabel('No of listing')
axes[0].set_ylabel('No of listing')

# set Plot title
plt.suptitle('Count Plot of Cities VS Point of Contact')

```

```python
fig, axes = plt.subplots(2, 3, figsize=(20,20))

sns.barplot(data=rent_data, x='City', ax=axes[0,0], hue='City')
sns.scatterplot(data=rent_data,x='Size', y='Rent', ax=axes[0,1])
sns.barplot(data=rent_data, x= 'Furnishing Status', y='Rent', ax=axes[1, 0])
sns.boxenplot(data=rent_data, x='City',y='Rent',ax=axes[1, 1], hue='City')
sns.histplot(data=rent_data[rent_data['Rent'] < 60000], x='Rent', ax=axes[0, 2])
sns.violinplot(data=rent_data[rent_data['Rent'] < 60000], x='City', y='Rent', ax=axes[1, 2], hue='City')
```
